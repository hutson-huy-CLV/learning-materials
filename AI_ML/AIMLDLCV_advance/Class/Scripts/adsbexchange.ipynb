```python
import pandas as pd
import json
import requests

base_url = 'https://samples.adsbexchange.com/readsb-hist/2024/07/01/{:06d}Z.json.gz'

# Define the start and end times (in seconds) and the step size
start_time = 0

# end_time = 24 * 60 * 60  # 24 hours in seconds
end_time = 50 #### uncomment the above to get 24 hours of data

step_size = 5  # Step size in seconds

# List to hold all the data
all_data = []

# Loop through the range and process each URL
for time in range(start_time, end_time, step_size):
    url = base_url.format(time)
    print(f"Processing URL: {url}")  # Print the URL for reference

    try:
        # Fetch the data from the URL
        r = requests.get(url, allow_redirects=True)
        if r.status_code == 200:
            data = json.loads(r.content)
            # Append the data to the list
            all_data.append(data)
        else:
            print(f"Failed to fetch data from {url} with status code: {r.status_code}")
    except Exception as e:
        print(f"Error fetching data from {url}: {e}")

# Save the collected data into a single JSON file
with open('combined_data.json', 'w') as outfile:
    json.dump(all_data, outfile)

print("Data successfully saved to combined_data.json")
```

```python
import json

# Load the data from the JSON file
with open('combined_data.json', 'r') as infile:
    all_data = json.load(infile)

data_aircraft = pd.concat([pd.DataFrame(data['aircraft']) for data in all_data], ignore_index=True)
```
